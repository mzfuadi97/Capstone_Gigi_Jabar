# Business Understanding

Rendahnya tingkat kesadaran tentang pentingnya merawat gigi pada usia dini di Jawa Barat merupakan masalah serius yang perlu segera diatasi. Upaya edukasi dan promosi kesehatan yang berkelanjutan sangat diperlukan untuk meningkatkan kesadaran masyarakat dan mencegah terjadinya penyakit gigi dan mulut pada anak-anak. 

Survei SKRT-SKN mengungkap, penyakit gigi dan mulut menduduki urutan pertama dari 10 penyakit yang dikeluhkan masyarakat dengan prevalensi 61 persen. Kelompok usia sekolah menjadi kalangan yang paling rentan terkena masalah kesehatan gigi dan mulut.
Faktor tersebut,  dikarenakan kurangnya akses terhadap pelayanan kesehatan gigi yang berkualitas, rendahnya tingkat kesadaran tentang pentingnya perawatan gigi, dan kurangnya jumlah tenaga kesehatan yang terlatih, menjadi penyebab utama. 

Oleh karena itu, perlu analisa lebih lanjut terkait perkembangan penyakit gigi dan mulut di daerah provinsi Jawa Barat.

### Preprocessing Data

#### Hasil Survey SKN-SRT 2022

```python
data = {
    'Nama Penyakit': ['Masalah gigi dan mulut', 'Hipertensi', 'Penyakit sendi', 'Penyakit pernapasan', 'Penyakit pencernaan', 'Diabetes', 'Penyakit jantung', 'Kanker', 'Stroke', 'Penyakit ginjal'],
    'Angka Prevalensi': [61.0, 34.0, 28.0, 27.0, 26.0, 15.0, 12.0, 8.0, 7.0, 6.0]
}

hasil_SKRT_SKN = pd.DataFrame(data)
hasil_SKRT_SKN['Angka Prevalensi'] = hasil_SKRT_SKN['Angka Prevalensi'] * 282000
hasil_SKRT_SKN.to_csv("dataset/dataset_clean/hasil_survei_clean.csv")
hasil_SKRT_SKN
```

##### Jumlah Tenaga Rumah Sakit dan Puskesmas dalam semua jenis

```python
df_jmlh_tenaga['Jenis Tenaga'] = df_jmlh_tenaga['Jenis Tenaga'].str.title()
df_jmlh_tenaga['tahun'] = df_jmlh_tenaga['tahun'].astype(str)
df_jmlh_tenaga['Jenis Tenaga'].value_counts()
```

```python
df_longlan = pd.read_csv("dataset/dataset_jds/diskominfo_longlan_jabar.csv")
df_longlan = df_longlan[['bps_kota_nama','latitude','longitude']]
df_longlan['bps_kota_nama'] = df_longlan['bps_kota_nama'].str.title()
df_longlan['bps_kota_nama_shp'] = df_longlan['bps_kota_nama'].str.replace('Kabupaten', '')
df_longlan['bps_kota_nama_shp'] = df_longlan['bps_kota_nama_shp'].str.replace(' ', '')
df_longlan = df_longlan.dropna()
df_longlan.to_csv("dataset/dataset_clean/jabar_longlan_clean.csv")
df_longlan
```

#### Rasio Tenaga Gigi / 100.000 penduduk

```python
rasio_jmlh_tenaga = pd.read_csv("dataset/dataset_gbg/rasio_tenaga.csv")
rasio_jmlh_tenaga['tahun'] = rasio_jmlh_tenaga['tahun'].astype(str)
rasio_jmlh_tenaga['jenis_tenaga'].value_counts()
```

```python
# Normalisasi jenis_tenaga
rasio_jmlh_tenaga.loc[rasio_jmlh_tenaga['jenis_tenaga'] == "Keteknisian Medis", 'jenis_tenaga'] = "Keteknisan Medis"
rasio_jmlh_tenaga.loc[rasio_jmlh_tenaga['jenis_tenaga'] == "Perawat Gigi", 'jenis_tenaga'] = "Perawat"
rasio_jmlh_tenaga.loc[rasio_jmlh_tenaga['jenis_tenaga'] == "SKM", 'jenis_tenaga'] = "Kesehatan Masyarakat"
rasio_jmlh_tenaga.loc[rasio_jmlh_tenaga['jenis_tenaga'] == "Tenaga Teknis Kefarmasian", 'jenis_tenaga'] = "Kefarmasian"
rasio_jmlh_tenaga.loc[rasio_jmlh_tenaga['jenis_tenaga'] == "Sanitarian", 'jenis_tenaga'] = "Keterafian Fisik"
rasio_jmlh_tenaga.loc[rasio_jmlh_tenaga['jenis_tenaga'] == "Keterapian Fisik", 'jenis_tenaga'] = "Keterafian Fisik"
rasio_jmlh_tenaga.loc[rasio_jmlh_tenaga['jenis_tenaga'] == "Apoteker", 'jenis_tenaga'] = "Kefarmasian"
rasio_jmlh_tenaga.to_csv("dataset/dataset_clean/rasio_jmlh_tenaga_clean.csv")
rasio_jmlh_tenaga['jenis_tenaga'].value_counts()
```

```python
df_filtered = jmlh_case_gigi[['tahun', 'Jumlah Kasus Gigi', 'Jumlah Kasus dirujuk']]

df_subset = jmlh_case_gigi[['tahun', 'Jumlah Kasus Gigi', 'Jumlah Kasus dirujuk']]
df_subset_grouped = df_subset.groupby('tahun').sum().reset_index()

# Menggambarkan bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='tahun', y='Jumlah Kasus Gigi', data=df_subset_grouped, color='blue', label='Jumlah Kasus Gigi')
sns.barplot(x='tahun', y='Jumlah Kasus dirujuk', data=df_subset_grouped, color='orange', label='Jumlah Kasus dirujuk')

plt.title('Perkembangan Kasus Gigi dan Kasus dirujuk Setiap Tahun')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Kasus')
for index, value in enumerate(df_subset_grouped['Jumlah Kasus Gigi']):
    plt.text(index, value + 1000, str(value), ha='center', va='bottom', color='blue')

for index, value in enumerate(df_subset_grouped['Jumlah Kasus dirujuk']):
    plt.text(index, value + 1000, str(value), ha='center', va='bottom', color='orange')

plt.legend()
plt.show()
```
