# Business Understanding

Rendahnya tingkat kesadaran tentang pentingnya merawat gigi pada usia dini di Jawa Barat merupakan masalah serius yang perlu segera diatasi. Upaya edukasi dan promosi kesehatan yang berkelanjutan sangat diperlukan untuk meningkatkan kesadaran masyarakat dan mencegah terjadinya penyakit gigi dan mulut pada anak-anak. 

Isu stunting menjadi perhatian serius dalam konteks kesehatan masyarakat khususnya di Jawa Barat. Stunting menyebabkan efek yang serius terhadap kesehatan dan perkembangan anak-anak. Gigi dibentuk saat janin berusia 4 bulan dalam kandungan. Sehingga, ibu hamil sangat dianjurkan untuk mengkonsumsi cukup suplemen fluor dan kalsium untuk menunjang pertumbuhan dan perkembangan gigi anak nanti. Seperti diketahui bahwa kondisi stunting dan karies gigi memiliki hubungan timbal balik. Oleh karena itu, untuk mencegah stunting kita perlu melakukan pemenuhan gizi seimbang seimbang dengan menjaga kondisi gigi geligi, sehingga anak dapat makan dengan baik. Sebaliknya, mencegah karies gigi dapat kita lakukan dengan pemenuhan gizi seimbang terutama untuk mendapatkan fungsi saliva optimal.

Oleh karena itu, perlu analisa lebih lanjut terkait perkembangan penyakit gigi dan mulut di daerah provisni Jawa Barat.

##### Dataset Stunting Jawa Barat

```python
df_stunting = pd.read_csv("dataset/dataset_jds/stunting_kab_tahun.csv")
df_stunting = df_stunting[['nama_kabupaten_kota','persentase_balita_stunting','tahun']]
df_stunting['nama_kabupaten_kota'] = df_stunting['nama_kabupaten_kota'].str.title()
df_stunting['tahun'] = df_stunting['tahun'].astype(str)
df_stunting.to_csv("dataset/dataset_clean/df_stunting_clean.csv")
df_stunting
```

```python
df_stunting_cnt = pd.read_csv("dataset/dataset_jds/cnt_stunting.csv")
df_stunting_cnt = df_stunting_cnt[['nama_kabupaten_kota','jumlah_balita_stunting','tahun']]
df_stunting_cnt['nama_kabupaten_kota'] = df_stunting_cnt['nama_kabupaten_kota'].str.title()
df_stunting_cnt['tahun'] = df_stunting_cnt['tahun'].astype(str)
df_stunting_cnt.to_csv("dataset/dataset_clean/cnt_stunting_clean.csv")
df_stunting_cnt
```

```python
#### Korelasi antara kasus gigi dengan jumlah stanting
df_merged = pd.merge(df_stunting_cnt, jmlh_case_gigi, on=['tahun', 'nama_kabupaten_kota'], how='inner')
df_merged = df_merged[['nama_kabupaten_kota','jumlah_balita_stunting','tahun','Jumlah Kasus Gigi']]
df_merged = df_merged[df_merged['Jumlah Kasus Gigi'] != 0]
df_merged
```

```python
sns.scatterplot(x='jumlah_balita_stunting', y='Jumlah Kasus Gigi', data=df_merged)
plt.title('Scatter Plot - Korelasi antara jumlah_balita_stunting dan Jumlah Tumpatan Gigi Tetap')
plt.xlabel('Jumlah Balita Stunting')
plt.ylabel('Jumlah Tumpatan Gigi Tetap')
plt.show()
```

```python
correlation = df_merged['jumlah_balita_stunting'].corr(df_merged['Jumlah Kasus Gigi'])
print(f'Correlation between jumlah_balita_stunting and Jumlah Kasus Gigi: {correlation}')
```

Dari hasil korelasi antara kasus gigi dengan stunting tidak begitu kuat dalam hubungan, sehingga akan mengganti hipotesa saya
